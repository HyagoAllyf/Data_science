```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Carregando os dados
base = pd.read_csv('C:/Users/hyago/documentos/Meusprojetos/python/data_science/trees.csv')
base.shape
```

```python
# dados
base.head()
```

```python
# Criação do histograma, considerando somente o segundo atributo da base de dados e com duas divisões(bins)
# A variável 'h' armazena as faixas de valores de Height

h = np.histogram(base.iloc[:,1], bins = 6)
h
```

```python
# Visualização do histograma com 6 divisões(bins)
plt.hist(base.iloc[:,1], bins = 10)
plt.title('Árvores')
plt.ylabel('Frequência')
plt.xlabel('Altura')
```

```python
# Histograma com a linha de distribuição de frequência, com 6 divições(bins)
# kde = linhas de densidade

sns.histplot(base.iloc[:,1], kde=False, bins=6, color='blue').set(title='Árvores')
```

```python
# densidade, método kdeplot para densidade
sns.kdeplot(base.iloc[:,1], color='blue').set(title='Árvores')
```

```python
#densidade e histograma
sns.histplot(base.iloc[:,1], kde=True, bins=6, color='blue').set(title='Árvores')
```

## Gráfico de Dispersão

Muito bom para comparar duas variáveis númericas

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de linhas considerando o volume e o atributo "girth" - sem marker
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
#fit_reg linha de tendência
sns.regplot(x=base.Girth, y=base.Volume, data = base, x_jitter=0.3, fit_reg = True)
```

```python
# nova base de dados
base = pd.read_csv('C:/Users/hyago/documentos/Meusprojetos/python/data_science/co2.csv')
base.head()
```

```python
# criando duas variáveis par cada atributo (x=conc e y = uptake)
x = base.conc
y = base.uptake
```

```python
# Retorna os valores únicoas do atributo "treatment"

unicos = list(set(base.Treatment))
unicos
```

```python
# Percorre cada tipo de tratamento (chilled e nonchilled) e cria um gráfico de dispersão

for i in range(len(unicos)):
    indice = base.Treatment == unicos[i]
    plt.scatter(x[indice], y[indice], label = unicos[i])
plt.legend(loc = 'lower right')
```

## Gráfico em 3D

```python
base = pd.read_csv('C:/Users/hyago/documentos/Meusprojetos/python/data_science/Data_science/orchard.csv')
base
```

```python
# Crianfo o gráfico 3D, indicano o atributo projection = '3d' e passando três atributo (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1, 1, 1, projection = '3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
```
